```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('NILAM.jpeg')
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
median_filtered_image = np.zeros_like(image_rgb)
for i in range(3):
    median_filtered_image[:, :, i] = cv2.medianBlur(image_rgb[:, :, i], 3)
```

# menampilkan citra asli dan filter median

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 7))
ax = axes.ravel()

ax[0].imshow(image_rgb)
ax[0].set_title('Image')
ax[0].axis('on')

ax[1].imshow(median_filtered_image)
ax[1].set_title('Median Filtered Image')
ax[1].axis('on')

plt.show()
```

```python
grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
mean_filtered_image = cv2.blur(grayscale_image, (5, 5))
```

# MENAMPILKAN CITRA GRAYSCALE DAN CITRA FILTER MEAN

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 6))
ax = axes.ravel()

ax[0].imshow(gray_image, cmap='gray')
ax[0].set_title('Image (Grayscale)')
ax[0].axis('on')

ax[1].imshow(mean_filtered_image, cmap='gray')
ax[1].set_title('Mean Filtered Image')
ax[1].axis('on')

plt.show()
```

# semua citra

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 12))

axs[0, 0].imshow(image_rgb)
axs[0, 0].set_title("image")
axs[0, 0].axis('on')

axs[0, 1].imshow(median_filtered_image)
axs[0, 1].set_title("Median Filtered Image")
axs[0, 1].axis('on')

axs[1, 0].imshow(gray_image, cmap='gray')
axs[1, 0].set_title("Image (Grayscale)")
axs[1, 0].axis('on')

axs[1, 1].imshow(mean_filtered_image, cmap='gray')
axs[1, 1].set_title("Mean Filtered Image")
axs[1, 1].axis('on')

plt.show()
```
